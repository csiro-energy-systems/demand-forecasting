```python
import pandas as pd
```

```python
# data from Electric Vehicle Council 2023 report
evc_data = {
    "Year": ["2011-12-31", "2012-12-31", "2013-12-31", "2014-12-31", "2015-12-31", "2016-12-31", "2017-12-31", "2018-12-31", "2019-12-31", "2020-12-31", "2021-12-31", "2022-12-31", "2023-06-30"],
    "BEV Sales": [49, 173, 191, 371, 759, 668, 1208, 1053, 5292, 5215, 17293, 33416, 43092],
    "PHEV Sales": [0, 80, 102, 951, 1012, 701, 1076, 1163, 1426, 1685, 3372, 5937, 3532],
    "Total EV Sales": [49, 253, 293, 1322, 1771, 1369, 2284, 2216, 6718, 6900, 20665, 39353, 46624],
    "EV Market Share Percent": [0.00, 0.02, 0.02, 0.12, 0.15, 0.12, 0.19, 0.21, 0.65, 0.78, 1.95, 3.81, 8.40]
}

# Creating DataFrame
evc_df = pd.DataFrame(evc_data)

# Convert to sales per year (i.e., take care of that last column consistently)

frac_of_year = [int(yr_str.split("-")[1])/12 for yr_str in evc_df["Year"]]
evc_df["BEV Sales Per Year"] = evc_df["BEV Sales"] / frac_of_year
evc_df["PHEV Sales Per Year"] = evc_df["PHEV Sales"] / frac_of_year
evc_df["Total EV Sales Per Year"] = evc_df["Total EV Sales"] / frac_of_year
evc_df['BEV Sales Percent'] = evc_df['BEV Sales Per Year'] / evc_df['Total EV Sales Per Year'] * evc_df['EV Market Share Percent']
evc_df["PHEV Sales Percent"] = evc_df["PHEV Sales Per Year"] / evc_df["Total EV Sales Per Year"] * evc_df["EV Market Share Percent"]
evc_df.drop(columns=["BEV Sales", "PHEV Sales", "Total EV Sales"], inplace=True)
#evc_df.to_csv("../data/processed/evc_df.csv", index=False)
evc_df
```

```python
evc_df.set_index("Year", inplace=True)
```

```python
# transpose df and use year as index
evc_df_t = evc_df.transpose()
display(evc_df_t)
evc_df_t.reset_index(inplace=True)
display(evc_df_t)

# create a new row with "Year" equal to "Total" and is "Total EV Sales Per Year" / "EV Market Share Percent" * 100
```

```python

display(evc_df_t)
```

```python
# Create a Total sales row
ev_pc = evc_df_t.loc[0].values[1:]
ev_sa = evc_df_t.loc[3].values[1:]
to_sa = ev_sa / ev_pc * 100
new_row = ["All"] + list(to_sa)
evc_df_t.loc[6] = new_row

# drop rows containing Percent
evc_df_t = evc_df_t[~evc_df_t["index"].str.contains("Percent")]
evc_df_t = evc_df_t[~evc_df_t["index"].str.contains("Total")]
display(evc_df_t)

def map_type(s):
    if "BEV" in s:
        return "Electric"
    elif "PHEV" in s:
        return "Plugin hybrid"
    elif "All" in s:
        return "All"
    else:
        return "Total"

# split the index column into two columns named "Vehicle Type" and "Fuel Type"
evc_df_t["Fuel Type"] = [map_type(s) for s in evc_df_t["index"]]
display(evc_df_t)

# drop index column
evc_df_t.drop(columns=["index"], inplace=True)
display(evc_df_t)

# rename index
evc_df_t.columns.name = "index"
display(evc_df_t)

# move last columt to first column
cols = evc_df_t.columns.tolist()    
cols = cols[-1:] + cols[:-1]
evc_df_t = evc_df_t[cols]

# reset index
evc_df_t.reset_index(inplace=True)
evc_df_t.drop(columns=["index"], inplace=True)

# save to csv
evc_df_t.to_csv("../data/processed/evc_df.csv", index=False)

display(evc_df_t)
```
